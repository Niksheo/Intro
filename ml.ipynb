```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
file=sns.load_dataset('tips')
```

```python
file
```

```python
file.to_csv('tips.csv')
```

```python
x=file[['total_bill','size']]#indipendent variable
```

```python
x
```

```python
y=file['tip']#dipendent variable
```

```python
y
```

```python
sns.scatterplot(x='total_bill',y='tip',data=file)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,random_state=2)
```

```python
x
```

```python
x_train
```

```python
x_test
```

```python
y_test
```

```python
regr=LinearRegression()
```

```python
regr.fit(x_train,y_train)
```

```python
y_pred=regr.predict(x_test)
```

```python
y_pred
```

```python
regr.predict([[100,5]])
```

```python
from sklearn.metrics import mean_absolute_error,r2_score,mean_squared_error
```

```python
print ('MAE',mean_absolute_error(y_test,y_pred))
```

```python
print ('mae',mean_squared_error(y_test,y_pred))
```

```python
import numpy as ny
```

```python
print('rmse',ny.sqrt(mean_squared_error(y_test,y_pred)))
```

```python
print('r_2 score',r2_score(y_test,y_pred))
```
