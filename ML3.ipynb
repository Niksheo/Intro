```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
df=sns.load_dataset('mpg')
```

```python
df
```

```python
df.isnull().sum()
```

```python
df=df.dropna()
```

```python
x=df[['cylinders','displacement','horsepower','weight','acceleration']]
```

```python
x
```

```python
y=df['mpg']
```

```python
y
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.75,random_state=1)
```

```python
LR=LinearRegression()
```

```python
LR.fit(x_train,y_train)
```

```python
y_pridict=LR.predict(x_test)
```

```python
y_pridict
```

```python
y_test
```

```python
mpg=LR.predict([[4,140.0,86.0,2790,15.6]])
print(mpg)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
print('r2_score',r2_score(y_test,y_pridict))
```

```python
print('MAE',mean_absolute_error(y_test,y_pridict))
```

```python
print('RMAE',mean_squared_error(y_test,y_pridict))
```

```python
import numpy as ny
```

```python
print('mae',ny.sqrt(mean_squared_error(y_test,y_pridict)))
```
